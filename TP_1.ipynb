# Clustering k-Means

```python
from sklearn import cluster
from sklearn import metrics
import time
from sklearn import cluster
from sklearn import metrics
```

```python
path = './artificial/'
databrut = arff.loadarff(open(path+"xclara.arff", 'r'))
datanp = np.array([[x[0],x[1]] for x in databrut[0]])
f0 = datanp[:,0] # tous les élements de la première colonne
f1 = datanp[:,1] # tous les éléments de la deuxième colonne

plt.scatter(f0, f1, s=8)
plt.title("Donnees initiales")
plt.show()
```

```python
print("Appel KMeans pour une valeur de k fixée (données init)")
tps1 = time.time()
k=3
model_km = cluster.KMeans(n_clusters=k, init='k-means++')
model_km.fit(datanp)
tps2 = time.time()
labels_km = model_km.labels_
# Nb iteration of this method
iteration = model_km.n_iter_

# Résultat du clustering
plt.scatter(f0, f1, c=labels_km, s=8)
plt.title("Données (init) après clustering")
plt.show()
print("nb clusters =",k,", nb iter =",iteration, ", runtime = ", round((tps2 - tps1)*1000,2),"ms")
#print("labels", labels_km)
# Some evaluation metrics
# inertie = wcss : within cluster sum of squares
inert = model_km.inertia_
silh = metrics.silhouette_score(datanp, model_km.labels_, metric='euclidean')
print("Inertie : ", inert)
print("Coefficient de silhouette : ", silh)
```

```python
##### Evaluation de k-means en fonction du paramètre k, le nombre de cluster fixé. Pour chaque valeur de k (de 2 à 10)
#### nous calculerons l'inertie et le coefficient de silhouette. 

K=np.arange(2, 11, 1)
tab_inert=[]
tab_silh=[]
tab_time=[]


for k in K:
    tps1 = time.time()
    model_km = cluster.KMeans(n_clusters=k, init='k-means++')
    model_km.fit(datanp)
    tps2 = time.time()
    iteration = model_km.n_iter_
    inert = model_km.inertia_
    silh = metrics.silhouette_score(datanp, model_km.labels_, metric='euclidean')
    tab_inert.append(inert)
    tab_silh.append(silh)
    tab_time.append(round((tps2 - tps1)*1000,2))
    print("k=", k, ",nb iter :", iteration, ",runtime :",round((tps2 - tps1)*1000,2),"ms, Inertie : ",inert, "silh :", silh)
    
```

```python
plt.plot(K, tab_inert)
plt.xlabel("nb cluster")
plt.ylabel("inertie")
plt.title("inertie vs k")
plt.show()
```

```python
plt.plot(K, tab_silh)
plt.xlabel("nb cluster")
plt.ylabel("silhouet coefficient")
plt.title("silhouet coefficient vs k")
plt.show()
```

```python
plt.plot(K, tab_time)
plt.xlabel("nb cluster")
plt.ylabel("time (ms)")
plt.title("time vs k")
plt.show()
```

```python

#To find elbow with inertia
def find_elbow(X, K, alpha_k, inertia_0):
    scaled_inertia=X/inertia_0+alpha_k*K
    k=K[np.where(scaled_inertia==min(scaled_inertia))]
    return scaled_inertia, k    
```
